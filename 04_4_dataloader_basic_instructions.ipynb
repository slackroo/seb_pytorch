# Pipeline for creating a Dataloader

1) Create Dataset
2) Instantiate Dataset
3) Intatiate Dataloader
4) Test/use Dataloader

```python
import pandas as pd
import torch
import os
from PIL import Image
from torch.utils.data import Dataset, DataLoader
```

```python
# 1) Create Dataset

class MyDataset(Dataset):

    def __init__(self, csv_path, img_dir, transform=None):
        # set up attributes for get_items method

        df = pd.read_csv(csv_path)
        self.img_dir = img_dir
        self.transform = transform

        # Based on dataFrame columns
        self.img_names = df['filepath']
        self.labels = df['label']


    def __len__(self):
        return self.labels.shape[0]

    def __getitem__(self, idx):
        # Define how to load a single record (training example, test example)

        img = Image.open(os.path.join(self.img_dir, self.img_names[idx]))
        label = self.labels[idx]

        if self.transform:
            img = self.transform(img)

        return img, label



```
